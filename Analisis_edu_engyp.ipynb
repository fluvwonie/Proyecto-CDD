# **Analisis de los datos**

```python
import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns
```

```python
import pandas as pd 
df=pd.read_csv('https://raw.githubusercontent.com/fluvwonie/World-/refs/heads/main/base_de_datos_limpia.csv')
df
```

Mostrar los datos estadisticos

```python
df.describe().T
```

```python
df.head(10)
```

```python
#Comprobar valores unicos en cada columna 
for column in df.columns:
    print(column, df[column].nunique())
```

```python
#Calcular el promedio de las calificaciones

# Definir las columnas relacionadas con las materias
subject_columns = ['Subject_1', 'Subject_2', 'Subject_3', 'Subject_4', 'Subject_5',
                   'Subject_6', 'Subject_7', 'Subject_8', 'Subject_9', 'Subject_10']

# Calcular el promedio de todas las materias para cada estudiante
df['Average_Grade'] = df[subject_columns].mean(axis=1)

# Agrupar por 'Father Degree' y calcular el promedio de calificaciones
average_grades_by_father_degree = df.groupby('Father Degree')['Average_Grade'].mean()

# Mostrar los resultados
print(average_grades_by_father_degree)

```

```python
#Filtrado del nombre 
df[df['Student Name']=='Michael Davis']

```

```python
nombres_similares = df['Student Name'].value_counts()[:30]

nombres_similares
```

```python

nombres_frecuentes = df['Student Name'].value_counts().nlargest(20)


colors = plt.cm.RdBu(np.linspace(0, 1, len(nombres_frecuentes)))


nombres_frecuentes.plot(kind='barh', color=colors)


plt.title('Top 20 nombres mÃ¡s Frecuentes', fontsize=16)
plt.xlabel('Numero de frecuencia', fontsize=12)
plt.ylabel('Nombre del estudiante', fontsize=12)

plt.grid(True, which='both', axis='x', linestyle='--', linewidth=0.7, alpha=0.7)


plt.gca().invert_yaxis()


plt.show()
```

 Analisis de la edad del estudiante y del tipo de educacion 

```python
df['Student Age'].value_counts()
```

```python

student_counts = df['Student Age'].value_counts()

# GrÃ¡fico de pastel
color = sns.color_palette('BrBG')

plt.figure(figsize=(6,6))
plt.pie(student_counts, labels=student_counts.index, autopct='%1.1f%%', startangle=90, colors=color)
plt.title('DistribuciÃ³n por Nombre del Estudiante')
plt.show()
```

```python

edu_counts = df['Education Type'].value_counts()

# GrÃ¡fico de pastel
color = sns.color_palette('PiYG')

plt.figure(figsize=(6,6))
plt.pie(edu_counts, labels=edu_counts.index, autopct='%1.1f%%', startangle=90, colors=color)
plt.title('DistribuciÃ³n del Tipo de Educacion')
plt.show()
```

```python
df['Student Age'].groupby(df['Education Type']).value_counts()
```

```python


sns.histplot(data=df, x='Student Age',hue ='Education Type', multiple="stack",palette="PuBu",kde=True, bins=30)

# Mostrar el grÃ¡fico
plt.title('DistribuciÃ³n del numero de estudiantes en cada Edad segun el tipo de educacion')
plt.xlabel('Edad del estudiante')
plt.ylabel('Numero de estudiantes')
plt.show()
```
